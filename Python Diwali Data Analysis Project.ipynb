```python
sns.countplot(x='Gender', data=df, hue='Gender',palette= 'viridis')
```

```python
ax = sns.countplot(x = 'Gender',data = df, hue='Gender', palette='plasma')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')
sns.barplot(x='Gender', y='Amount', data=sales,hue='Gender', palette='BuPu')
```

```python
dx =sns.countplot(x= 'Age Group',hue= 'Gender', data=df)

for bars in dx.containers:
    dx.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group', palette='GnBu')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(12)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders', hue='State', palette='magma')
```

```python
sales_amt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount').head(11)
sns.barplot(x='State', y='Amount', data= sales_amt, hue='State', palette='plasma')
```
